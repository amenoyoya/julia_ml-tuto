# 微積分

## 微分

関数 $y=f(x)$ が与えられた時に $x=a$ から $x=b$ への値の変化に注目する

この間の $f(x)$ の増分の $x$ の増分に対する割合、

$$
\frac{f(b) - f(a)}{b - a}
$$

を $x$ が $a$ から $b$ に変化するときの **平均変化率** と呼ぶ

これはつまり、点 $(a, f(a))$ と点 $(b, f(b))$ を結ぶ直線の傾きである

![rate_of_change.drawio.png](./img/rate_of_change.drawio.png)

$b$ を $a$ に近づけていくことで、点 $(a, f(a))$ と点 $(b, f(b))$ を通る直線は下図のように、点 $(a, f(a))$ における $y=f(x)$ の **接線** となる

![tangent.drawio.png](./img/tangent.drawio.png)

このときの平均変化率の極限は接線の傾きになる

この接線の傾きのことを $x=a$ における $f(x)$ の **微分係数** と呼び $f'(a)$ で表す

$$
f'(a) = \lim_{b\to{a}}\frac{f(b) - f(a)}{b - a}
$$

ここで $b-a=h$ とおくと、$b\rightarrow a$ のとき $h\rightarrow0$ なので、次のように定義することが出来る

$$
f'(a) = \lim_{h\to0}\frac{f(a+h) - f(a)}{h}
$$

一般には上式を微分係数の定義として用いることが多い

この極限値が存在する時、関数 $f(x)$ は $x=a$ で微分可能である、という

逆に、次の関数は $x=0$ で右極限と左極限が異なるため、微分不可能である

$$
\begin{align}
f(x) = |x| \\
\\
\lim_{h\to+0}\frac{|x+h| - |x|}{h} = 1 \\
\lim_{h\to-0}\frac{|x+h| - |x|}{h} = -1 \\
\end{align}
$$

ここまで $a$ を固定して考えてきたが、$a$ を動かしていくと $f'(a)$ は各点での $y=f(x)$ の接線の傾きを表す関数と考えることが出来る

これを $x$ の関数と考えて $f'(x)$ と記述し、$f(x)$ の **導関数** と呼ぶ

導関数の表記にはいくつかあり、

$$
f'(x),\ \frac{df}{dx}(x),\ \frac{d}{dx}f(x),\ \frac{dy}{dx}
$$

は全て同じものである

特に $\frac{d}{dx}$ という表記は、$x$ の関数に対して左に作用させることで導関数を意味するため、記述上便利なことが多い

導関数を求めることを **微分** すると呼ぶ

### 導関数の性質
定数関数 $f(x) = k$ の導関数は 0 となる

実際 $f(x+h) = f(x) = k$ であるため、$\frac{k}{dx}k = \lim_{h\to0}\frac{k-k}{h} = 0$ となる

導関数には、次のような性質がある

$$
\begin{align}
\frac{d}{dx}(f(x) + g(x)) &= \frac{d}{dx}f(x) + \frac{d}{dx}g(x) \\
\frac{d}{dx}(kf(x)) &= k\frac{d}{dx}f(x)
\end{align}
$$

### 多項式の微分
多項式の定理として以下の **二項定理** が知られている

$$
(a+b)^n = {}_nC_0a^n + {}_nC_1a^{n-1}b + {}_nC_2a{n-2}b^2 + \cdots + {}_nC_nb^n
$$

ここで、

$$
{}_nC_k = \frac{n!}{k!(n - k)!}
$$

であり、特に ${}_nC_0=1$, ${}_nC1=n$ となる

これを使って多項式の微分を考える

一般の多項式の微分を考えるには、導関数の性質より、$x^n$ の微分を考えれば十分で、後は定数倍と和で導関数を求めることが出来る

そのため、以下のように $x^n$ の微分を考えてみる

$$
\begin{align}
\frac{d}{dx}x^n &= \lim_{h\to0}\frac{(x+h)^n - x^n}{h} \\
&= \lim_{h\to0}\frac{x^n + nhx^{n-1} + {}_nC_2h^2x^{n-2} + \cdots + {}_nC_nh^n - x^n}{h} \quad (二項定理より) \\
&= \lim_{h\to0}\frac{nhx^{n-1} + {}_nC_2h^2x^{n-2} + \cdots + {}_nC_nh^n}{h} \\
&= \lim_{h\to0}(nx^{n-1} + {}_nC_2hx^{n-2} + \cdots + {}_nC_nh^{n-1}) \\
&= nx^{n-1} \quad (h が掛かっている部分は全て消える)
\end{align}
$$

つまり $\frac{d}{dx}x^n = nx^{n-1}$ となる

これはよく使うため、公式として覚えておくと良い

この公式を用いて具体的な関数として $x^4-2x^2+3$ の導関数を考えてみる

$$
\begin{align}
\frac{d}{dx}(x^4-2x^2+3) &= \frac{d}{dx}(x^4) + \frac{d}{dx}(-2x^2) + \frac{d}{dx}(3) \\
&= \frac{d}{dx}(x^4) - 2\frac{d}{dx}(x^2) + \frac{d}{dx}(3) \\
&= 4x^3 - 2\times 2x + 0 \\
&= 4x^3 - 4x
\end{align}
$$

### 積・商の微分と高階道関数
関数の積の微分については、次の式が成り立つ

$$
\frac{d}{dx}(f(x)g(x)) = f'(x)g(x) + f(x)g'(x)
$$

この式は以下により証明される

$$
\begin{align}
\frac{d}{dx}(f(x)g(x)) &= \lim_{h\to0}\frac{f(x+h)g(x+h) - f(x)g(x)}{h} \\
&= \lim_{h\to0}\frac{f(x+h)g(x+h) - f(x)g(x+h) + f(x)g(x+h) - f(x)g(x)}{h} \\
&= \lim_{h\to0}\frac{(f(x+h) - f(x))g(x+h) + f(x)(g(x+h) - g(x))}{h} \\
&= \lim_{h\to0}\frac{f(x+h) - f(x)}{h}\cdot\lim_{h\to0}g(x+h) + \lim_{h\to0}f(x)\cdot\lim_{h\to0}\frac{g(x+h) - g(x)}{h} \\
&= f'(x)g(x) + f(x)g'(x)
\end{align}
$$

また、関数の商の微分については、次の式が成り立つ

$$
\frac{d}{dx}\frac{f(x)}{g(x)} = \frac{f'(x)g(x) - f(x)g'(x)}{g(x)^2}
$$

上記を証明するには、まず $\frac{d}{dx}\left(\frac{1}{g(x)}\right)$ の計算を行う

$$
\begin{align}
\frac{d}{dx}\left(\frac{1}{g(x)}\right) &= \lim_{h\to0}\frac{\frac{1}{g(x+h)} - \frac{1}{g(x)}}{h}) \\
&= \lim_{h\to0}\frac{g(x) - g(x+h)}{hg(x)g(x+h)} \\
&= \lim_{h\to0}\frac{-(g(x+h) - g(x))}{h}\cdot\lim_{h\to0}\frac{1}{g(x)g(x+h)} \\
&= -\frac{g'(x)}{g(x)^2}
\end{align}
$$

次に、上式を $\frac{d}{dx}\frac{f(x)}{g(x)}$ に当てはめて積の微分の公式を使うと以下のようになる

$$
\begin{align}
\frac{d}{dx}\left(\frac{f(x)}{g(x)}\right) &= \frac{d}{dx}\left(f(x)\cdot\frac{1}{g(x)}\right) \\
&= f'(x)\cdot\frac{1}{g(x)} + f(x)\cdot\left(\frac{1}{g(x)}\right)' \\
&= f'(x)\cdot\frac{1}{g(x)} + f(x)\cdot\left(-\frac{g'(x)}{g(x)^2}\right) \\
&= \frac{f'(x)g(x) - f(x)g'(x)}{g(x)^2}
\end{align}
$$

関数 $y=f(x)$ の導関数 $f'(x)$ のさらに導関数をとることがある

それを **2階導関数** と呼び、$f''(x)$, $\frac{d^2}{dx^2}f(x)$, $\frac{d^2y}{dx^2}$ などの書き方をする

さらに一般的には微分を $n$ 回行うこともあり、$f(x)$ を $n$ 回微分したものを$n$階導関数と呼び、$f^(n)(x)$ または $\frac{d^n}{dx^n}f(x)$ などで表す

すべての関数が $n$ 回微分可能とは限らず、特に何度でも（無限回）微分可能な関数を **滑らか** であるという

例えば任意の多項式関数は滑らかである

実際、$n$次元を微分すると$n-1$次元になるため、何度か微分すると定数になり、定数は何度でも微分可能（永遠に 0 になり続ける）、すなわち滑らかである

### 合成関数の微分と逆関数の微分
$w$ についての関数 $y=f(w)$ と $x$ についての関数 $w=g(x)$ が与えられた時、関数 $y=f(g(x))$ は **合成関数** と呼ばれる

これはつまり $x$ に対して、関数 $g$ を作用させた後に $f$ を作用させるという関数である

$f(x)$ と $g(x)$ が微分可能である時、合成関数の微分について次の式が成り立つ

$$
\frac{dy}{dx} = \frac{dy}{dw}\cdot\frac{dw}{dx}
$$

次に逆関数について考えてみる

関数 $y=f(x)$ について、もしこの関数の値域にあるすべての $y$ に対して $y=f(x)$ となる $x$ が1つしかないという条件を満たすとすると、$y$ を1つ決めると $y=f(x)$ を満たす $x$ は1つ決まることになり、この対応関係も関数 $x=g(y)$ とみなすことが出来る

この関数 $x=g(y)$ を $f(x)$ の **逆関数** と呼び、$f^{-1}$ のように書くことがある

逆関数の性質としては以下が成り立つ

$$
\begin{align}
f^{-1}(f(x)) &= x \\
f(f^{-1}(x)) &= x
\end{align}
$$

### 指数関数と対数関数の微分
自然対数の底 $e$ の定義 $e = \lim_{n\to\infty}(1 + \frac{1}{n})^n$ において、$\frac{1}{n}=t$ とおくと、$n\to\infty$ のとき $t\to0$ なので、

$$
e = \lim_{t\to0}(1+t)^\frac{1}{t}
$$

となる

ここで $\lim_{h\to0}\frac{e^h-1}{h}$ の極限を計算したいとする

$e^h-1=u$ とおくと、$h\to0$ のとき $u\to0$ であり、$e^h=1+u$ より $h=\ln(1+u)$ となり、以下が導かれる

$$
\begin{align}
\lim_{h\to0}\frac{e^h-1}{h} &= \lim_{u\to0}\frac{u}{\ln(1+u)} \\
&= \lim_{u\to0}\frac{1}{\frac{1}{u}\ln(1+u)} \\
&= \lim_{u\to0}\frac{1}{\ln(1+u)^\frac{1}{u}} \\
&= \lim_{u\to0}\frac{1}{\ln{e}} \\
&= 1
\end{align}
$$

上記の結果を利用して指数関数 $y=e^x$ の微分を行う

$$
\begin{align}
\frac{d}{dx}e^x &= \lim_{h\to0}\frac{e^{x+h}-e^x}{h} \\
&= \lim_{h\to0}\frac{e^x(e^h-1)}{h} \\
&= e^x\cdot\lim_{h\to0}\frac{e^h-1}{h} \\
&= e^x
\end{align}
$$

上記のように指数関数の微分は、元の関数と同じになる

このことから、$e$ という数が微分において特別な意味を持つことが分かる

すなわち、$e$ は指数関数の微分の計算を簡単にすることが出来るということである

実際、$e$ の代わりに任意定数 $a$ ($a\gt0$) を底とする指数関数 $y=a^x$ の微分を行ってみると、

$$
\frac{d}{dx}a^x = e^{x\ln{a}}\cdot\ln{a} = a^x\ln{a}
$$

となり、あまりきれいな形にならない

次に、対数関数 $y=\ln{x}$ の微分を考えてみると、以下のようになる

$$
\begin{align}
\frac{d}{dx}\ln{x} &= \lim_{h\to0}\frac{\ln(x+h)-\ln{x}}{h} \\
&= \lim_{h\to0}\left(\frac{1}{h}\ln\frac{x+h}{x}\right) \\
&= \frac{1}{x}\lim_{h\to0}\left(\frac{x}{h}\ln(1+\frac{h}{x})\right) \\
&= \frac{1}{x}\lim_{h\to0}\ln\left(1+\frac{h}{x}\right)^\frac{x}{h}
\end{align}
$$

ここで $\frac{h}{x}=t$ とおくと、$h\to0$ のとき $t\to0$ であるため、

$$
\lim_{h\to0}\ln\left(1+\frac{h}{x}\right)^\frac{x}{h} = \lim_{t\to0}\ln(1+t)^\frac{1}{t}=\ln{e}=1
$$

となり、つまり、

$$
\frac{d}{dx}\ln{x} = \frac{1}{x}
$$

である

これも、対数の底を $e$ にしたことにより結果が簡潔になることが分かる

実際、一般の $a\gt0$ を底とする対数関数 $y=\log_ax$ の導関数を求めると、$\log_ax=\frac{\log_ex}{\log_ea}$ より、

$$
\frac{d}{dx}\log_ax = \frac{1}{x\ln{a}}
$$

という、やや複雑な形の関数となる

以上より、指数関数と対数関数で、$e$ という数が特別な意味を持つことが確認できた

逆に考えれば、指数・対数関数を微分した結果がシンプルになるように決められた数が $e$ であるとも言える

なお、関数 $y=\ln{x}$ の微分の話をしたときは $x\gt0$ であることを暗黙的に仮定していた

この関数は $x\gt0$ の定義域でしか定義されないためである

そこで $\ln|x|$ の微分を考えてみる

この関数は $x\neq0$ で定義されているため、

$$
\ln|x| = \left\{
\begin{array}{cc}
\ln{x} & (x\gt0) \\
\ln(-x) & (x\lt0)
\end{array}
\right.
$$

となる

$x\gt0$ の場合はすでに計算済みのため、ここでは $x\lt0$ の範囲での微分を考える

$$
\begin{align}
\frac{d}{dx}\ln(-x) &= \frac{1}{-x}\times(-1) \quad (\because 合成関数の微分) \\
&= \frac{1}{x}
\end{align}
$$

したがって、$x\gt0$ でも $x\lt0$ でも等しく導関数は $\frac{1}{x}$ となり、

$$
\frac{d}{dx}\ln|x| = \frac{1}{x}
$$

を示すことができた

### 関数の増減と極大・極小
導関数の応用例として、関数の増減を調べてグラフの概形を推測するというものがある

例えば次のような関数について考えてみる

$$
f(x) = x^3-3x^2-9x+5
$$

この導関数は以下のようになる

$$
f'(x) = 3x^2-6x-9 = 3(x+1)(x-3)
$$

上記は $x\lt-1$ では正になり、$x=-1$ では 0 に、$-1\lt{x}\lt3$ では負に、$x=3$ では 0 に、$x\gt3$ では正になる

これを表で表すと以下のようになる

| x | ... | -1 | ... | 3 | ... |
|:--:|:--:|:--:|:--:|:--:|:--:|
|f'(x)| + | 0 | - | 0 | + |
|f'(x)| ↗ | 極大 | ➘ | 極小 | ↗ |

導関数とは接線の傾きを表していたため、これが正であることは、関数 $f(x)$ はその区間で増加していることになる

同様に、導関数が負の範囲では $f(x)$ は減少している

これを表に表すと、上記の表の3行目のようになり、この表を **増減表** と呼ぶ

特に導関数が 0 に等しくなるときが重要で、このとき $f(x)$ が極大または極小になることがある

- **極大**:
    - 十分に小さい近傍で最大であること
    - 極大のときの関数の値を **極大値** と呼ぶ
- **極小**:
    - 十分に小さい近傍で最小であること
    - 極小のときの関数の値を **極小値** と呼ぶ

```julia
using Plots

f(x) = x^3 - 3x^2 - 9x + 5
plot(f; x_lims=(-5, 5), y_lims=(-25, 25))
```

極大値と極小値をまとめて **極値** と呼ぶこともある

導関数が接線の傾きを表すことを思い出すと、極値をとる点で導関数が 0 になるのは直感的に分かる

一方、導関数が 0 であっても、必ずしも極大または極小とは限らないため注意が必要である

例えば $f(x)=x^3$ という関数を考えると、この導関数は $f'(x)=3x^2$ であるため、$x=0$ で $f'(x)=0$ となる

しかし $f'(x)=3x^2$ は $x\lt0$ でも $x\gt0$ でも正なので、$f(x)$ は $x=0$ の左でも右でも増加していることになり、$x=0$ では極値をとらない

```julia
f(x) = x^3
plot(f; x_lims=(-5, 5), y_lims=(-25, 25))
```

このように、$f'(x)=0$ を満たす $x$ で極大になるか、極小になるか、あるいは極値をとらないか調べるには、増減表を書くか2階導関数を調べる必要がある

#### 解答1
$$
f'(x) = 3x^2 - 3 = 3(x+1)(x-1)
$$

| x | ... | -1 | ... | 1 |
|:--:|:--:|:--:|:--:|:--:|
|f'(x)| + | 0 | - | 0 | + |
|f'(x)| ↗ | 極大 | ➘ | 極小 | ↗ |

以上より、極大値 $f(-1)=2$, 極小値 $f(1)=-2$

#### 解答2
$$
f'(x) = 12x^3 - 12x^2 - 72x = 12x(x-3)(x+2)
$$

| x | ... | -2 | ... | 0 | ... | 3 |
|:--:|:--:|:--:|:--:|:--:|:--:|:--:|
|f'(x)| - | 0 | + | 0 | - | 0 | + |
|f'(x)| ➘ | 極小 | ↗ | 極大 | ➘ | 極小 | ↗ |

増減表より、極小値は2つあり、最小値はそのうちのいずれかとなる

$$
\begin{align}
f(-2) &= -64 \\
f(3) &= -189
\end{align}
$$

より、最小値は $f(3) = -189$
